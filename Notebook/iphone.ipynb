# Top 10 highest rated Iphone on Flipkart in India

```python
df.head()
```

```python
highest_rated = df.sort_values(by= ["Star Rating"], ascending=False)
highest_rated = highest_rated.head(10)
```

```python
print(highest_rated[["Product Name"]])
```

```python
print(highest_rated)
```

# how many rating do the highest- rated iPhones on Flipkart have

```python
df.head()
```

```python
iphones = highest_rated['Product Name'].value_counts()
label = iphones.index
counts = highest_rated['Number Of Ratings']
figure = px.bar(highest_rated, x=label, y=counts )
figure.show()
```

# which iphone has the highest on flipkart

```python
df.head()
```

```python
iphones = highest_rated['Product Name'].value_counts()
label = iphones.index
counts = highest_rated['Number Of Reviews']
figure = px.bar(highest_rated, x=label, y=counts, title = "Number of Reviews for the Highest Rated iPhones")
figure.show()
```

# What is the relationship between the sale price of iphone and the number of rating on Flipkart?

```python
df.head()
```

```python
figure = px.scatter(
    data_frame=df,
    x="Number Of Ratings",
    y="Sale Price",
    size='Discount Percentage',
    color="Star Rating",
    trendline='ols',
    title='Relationship Between Number of Ratings and Sale Price'
)

figure.show()
```

- conclusion there is a negative relationship between the sale price of iphone and num of Rating

# relation discount percetage and the number of rating

```python
figure = px.scatter(
    data_frame=df,
    x="Number Of Ratings",
    y="Discount Percentage",
    size='Sale Price',
    color="Star Rating",
    trendline='ols',
    title='Relationship Between Discount Percentage and Number of Ratings'
)

figure.show()
```
